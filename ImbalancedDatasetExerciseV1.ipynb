```python
X = dataset.drop(columns=['Class'])
y = dataset['Class']
y = pd.DataFrame(y)
```

# Divide Training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=28 )
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
print(np.bincount(y_train.values.ravel()), 
      np.bincount(y_test.values.ravel()))
```

# Build the model

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix, precision_recall_fscore_support
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
pd.DataFrame(X_train_std).describe()
```

```python
svc_model=SVC()
svc_model.fit(X_train_std,y_train.values.ravel())
```

```python
from sklearn.metrics import accuracy_score
y_pred = svc_model.predict(X_test_std)
print("the accuracy of the model is ", accuracy_score(y_test, y_pred))
```

```python
print("The confusion matarix for the model is \n",
      confusion_matrix(y_test, y_pred))
```

```python
print("The precision recall vals for the model is \n",
      precision_recall_fscore_support(y_test, y_pred))
```
